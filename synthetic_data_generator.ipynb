```python
import pandas as pd
import random
from faker import Faker

fake = Faker()

# Define some example skill sets
skill_pool = [
    'Python', 'Data Science', 'Machine Learning', 'JavaScript', 'React', 'Node.js',
    'HTML', 'CSS', 'Django', 'Flask', 'Pandas', 'NLP', 'AWS', 'Docker', 'SQL'
]

# Helper to get random subset of skills
def random_skills(n=3):
    return ', '.join(random.sample(skill_pool, n))

# Generate Freelancers Data
num_freelancers = 50
freelancers = []

for i in range(num_freelancers):
    freelancer = {
        'Freelancer_ID': f'F{1000 + i}',
        'Name': fake.name(),
        'Skills': random_skills(random.randint(2, 5)),
        'Hourly_Rate': round(random.uniform(10, 100), 2),
        'Location': fake.country(),
        'Rating': round(random.uniform(3.5, 5.0), 1),
        'Completed_Projects': random.randint(5, 100)
    }
    freelancers.append(freelancer)

freelancers_df = pd.DataFrame(freelancers)
freelancers_df.to_csv('freelancers.csv', index=False)

# Generate Jobs Data
num_jobs = 30
jobs = []

for i in range(num_jobs):
    job = {
        'Job_ID': f'J{2000 + i}',
        'Title': fake.job(),
        'Required_Skills': random_skills(random.randint(2, 4)),
        'Budget': random.randint(300, 5000),
        'Duration_Days': random.randint(3, 30),
        'Posted_By': f'C{random.randint(100, 150)}'
    }
    jobs.append(job)

jobs_df = pd.DataFrame(jobs)
jobs_df.to_csv('jobs.csv', index=False)

print("✅ Synthetic CSVs generated: freelancers.csv and jobs.csv")
```
